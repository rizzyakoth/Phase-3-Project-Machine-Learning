# SYRIATEL CUSTOMER CHURN    
# Overview
In Syria, telecommunications services face a unique set of circumstances. The industry is highly regulated, with the government-owned Syrian Telecommunications Establishment (STE) maintaining a stronghold on fixed-line markets. This regulatory environment, combined with infrastructure disparities, particularly in remote areas relying on satellite communications, creates a complex landscape for telecom providers. Despite these challenges, SyriaTel strives to provide reliable and innovative services to its diverse customer base.

# Problem Statement    
SyriaTel, a telecommunications company, is experiencing a significant rate of customer churn, leading to a negative impact on revenue and customer satisfaction. The company is seeking a data-driven solution to predict customer churn accurately and implement proactive retention strategies. To address this challenge, the objective is to identify customers at risk of leaving and take targeted actions to retain them, ultimately reducing the overall churn rate.
# Objectives  
Our primary challenge lies in predicting and mitigating customer churn within the confines of a regulated and infrastructurally diverse telecom market. Specifically, to:
1.	Identify the key factors influencing customer churn by analyzing historical data.
2.	Develop a predictive model- Create a machine learning model to predict customer churn based on historical data.
3.	Enhance customer satisfaction by addressing specific issues contributing to churn.
4.	Implement proactive retention strategies- Utilize insights from the predictive model to develop and implement targeted retention strategies for at-risk customers.





# Data Understanding
The dataset used for this project has been retrieved from https://www.kaggle.com/datasets/becksddf/churn-in-telecoms-dataset/code as 'bigml_59c28831336c6604c800002a.csv'. We will:
1. Load the data
2. Display basic information about the dataset
3. Display statistical description of all numeric columns
4. Display the shape of the dataset
5. Display value count for our target variable 'churn'    

We will be using OOP(Object Oriented Programming),we will define the class ChurnDataProcessor and put all the attributes and methods mentioned above.

```python
import pandas as pd

class ChurnDataProcessor:
    def __init__(self, data_path):
        self.df = pd.read_csv(data_path)

    def display_dataset_info(self):
        # Display basic information about the dataset
        print("Dataset Information:")
        print(self.df.info())

    def describe_numeric_columns(self):
        # Display statistical description of numeric columns
        print("\nStatistical Description of Numeric Columns:")
        return self.df.describe()

    def display_dataset_shape(self):
        # Display the shape of the dataset
        print("\nDataset Shape:")
        print(self.df.shape)
        
    def display_categorical_value_counts(self, column):
        # Display value counts for a specific categorical column
        print(f"\nValue counts for {column}:")
        print(self.df[column].value_counts())
        
    def display_churn_value_counts(self):
        # Display value counts for the 'churn' column
        self.display_categorical_value_counts('churn')

# Load the data
data_path = 'bigml_59c28831336c6604c800002a.csv'

# Create an instance of ChurnDataProcessor
processor = ChurnDataProcessor(data_path)

# Display the head of the dataset
processor.df.head()
```

```python
# Display basic information about the dataset
processor.display_dataset_info()
```

* All  features are numeric other than: state, phone number, international plan and voice mail plan

# Data Preparation  
Prepare the data for EDA and modeling. We will check for: 
* Missing values
* Duplicate rows   

Then:
* Encode our categorical variables
* Normalize numerical features or variables

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
class ChurnDataCleaner(ChurnDataProcessor):
    def __init__(self, data_path):
        super().__init__(data_path)
    def check_missing_values(self):
        # Check for missing values in the dataset
        return self.df.isnull().sum()
    def check_duplicates(self):
        # Check for duplicates in the entire dataset
        duplicates = self.df.duplicated().sum()
        print("\nNumber of Duplicates in the Dataset:", duplicates)
    def encode_categorical_variables(self, categorical_columns):
        # One-hot encode categorical variables
        self.df = pd.get_dummies(self.df, columns=categorical_columns)
        return self.df
    def normalize_numeric_features(self):
        # Normalize numeric features using StandardScaler
        scaler = StandardScaler()
        numeric_columns = self.df.select_dtypes(include=['float64']).columns
        self.df[numeric_columns] = scaler.fit_transform(self.df[numeric_columns])
        print("Numeric features normalized.")

categorical_columns = ['international plan', 'voice mail plan']
# Create an instance of ChurnDataCleaner
cleaner = ChurnDataCleaner(data_path)
```

```python
# Check missing values
cleaner.check_missing_values()
```

```python
# Check for duplicates
cleaner.check_duplicates()
```

Our dataset seem to be okay since there are no missing values and duplicates.

```python
# One-hot encode specified categorical columns
cleaner.encode_categorical_variables(categorical_columns)
```

```python
# Normalize numeric features using StandardScaler scaling
cleaner.normalize_numeric_features()
```

# Data Analysis  
Here is whereby we use EDA to do our analysis:
* Identify patterns or correlations that may be indicative of 'churn' our target variable
* Visualize relationships between 


```python
import matplotlib.pyplot as plt
import seaborn as sns
import math  # Import the math module for ceil function
class ChurnDataAnalyzer(ChurnDataCleaner):
    def __init__(self, data_path):
        super().__init__(data_path)

    def identify_churn_patterns(self):
        # Identify patterns or correlations indicative of churn
        churn_corr = self.df.corr()['churn'].sort_values(ascending=False)
        return churn_corr
    def visualize_individual_features(self):
        #Visualize all numeric features on a single grid
        numeric_columns = self.df.select_dtypes(include=['float64', 'int64']).columns
        num_features = len(numeric_columns)

        # Determine the number of rows and columns for the grid
        num_rows = int(num_features / 2) + (num_features % 2)
        num_cols = 2

        plt.figure(figsize=(15, 5 * num_rows))

        # Plot each numeric feature on the grid
        for i, column in enumerate(numeric_columns, 1):
            plt.subplot(num_rows, num_cols, i)
            sns.histplot(data=self.df, x=column, kde=True, palette='coolwarm')
            plt.title(f'Distribution of {column}')
            plt.xlabel(column)
            plt.ylabel('Frequency')

        plt.tight_layout()
        plt.show()

    def visualize_relationships(self):
        # Visualize relationships between different features and the target variable (churn)
        plt.figure(figsize=(14, 10))
        sns.heatmap(self.df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
        plt.title('Correlation Heatmap')
        plt.show()
    def visualize_categorical_features(self, categorical_columns):
        # Visualize relationships between categorical features and the target variable (churn)
        plt.figure(figsize=(16, 8))
        for column in categorical_columns:
            plt.subplot(2, 2, categorical_columns.index(column) + 1)
            sns.countplot(x=column, hue='churn', data=self.df, palette='coolwarm')
            plt.title(f'{column} vs Churn')
        plt.tight_layout()
        plt.show()
    
    def visualize_area_code_and_churn(self):
        # Visualize the relationship between 'area code' and 'churn' using a count plot
        plt.figure(figsize=(10, 6))
        sns.countplot(x='area code', hue='churn', data=self.df, palette='coolwarm')
        plt.title('Count of Churn for Each Area Code')
        plt.xlabel('Area Code')
        plt.ylabel('Count')
        plt.show()
    def visualize_account_length_and_churn(self):
        # Visualize the relationship between 'account length' and 'churn' using a box plot
        plt.figure(figsize=(12, 6))
        sns.boxplot(x='churn', y='account length', data=self.df, palette='coolwarm')
        plt.title('Account Length vs. Churn')
        plt.xlabel('Churn')
        plt.ylabel('Account Length')
        plt.show()
    def drop_outliers_numeric_features(self, threshold=3):
        # Drop outliers from numeric features more than 'threshold' standard deviations away from the mean
        numeric_columns = self.df.select_dtypes(include=['float64', 'int64']).columns

        for column in numeric_columns:
            mean = self.df[column].mean()
            std_dev = self.df[column].std()

            lower_bound = mean - threshold * std_dev
            upper_bound = mean + threshold * std_dev

            self.df = self.df[(self.df[column] >= lower_bound) & (self.df[column] <= upper_bound)]

    #def drop_outliers_numeric_features(self):
        # Drop outliers from numeric features using the IQR method
        #numeric_columns = self.df.select_dtypes(include=['float64', 'int64']).columns

        #for column in numeric_columns:
           # Q1 = self.df[column].quantile(0.25)
           # Q3 = self.df[column].quantile(0.75)
            #IQR = Q3 - Q1

            #lower_bound = Q1 - 1.5 * IQR
            #upper_bound = Q3 + 1.5 * IQR

            #self.df = self.df[(self.df[column] >= lower_bound) & (self.df[column] <= upper_bound)]

# Dataset
data_path = 'bigml_59c28831336c6604c800002a.csv'

# Create an instance of ChurnDataAnalyzer
analyzer = ChurnDataAnalyzer(data_path)
```

```python
# Import necessary libraries
from sklearn.preprocessing import MinMaxScaler
class ChurnFeatureEngineering(ChurnDataAnalyzer):
    def __init__(self, data_path):
        super().__init__(data_path)
        
    def scale_numeric_features(self):
        # Scale numeric features using Min-Max scaling
        numeric_columns = self.df.select_dtypes(include=['float64', 'int64']).columns

        scaler = MinMaxScaler()
        self.df[numeric_columns] = scaler.fit_transform(self.df[numeric_columns])

    def create_international_plan_binary(self):
        # Create a binary feature indicating whether the customer has an international plan or not
        self.df['has_international_plan'] = (self.df['international plan'] == 'yes').astype(int)
    def create_has_voice_mail_plan_binary(self):
        # Create a binary feature indicating whether the customer has a voice mail plan or not
        self.df['has_voice_mail_plan'] = (self.df['voice mail plan'] == 'yes').astype(int)

# Create an instance of ChurnFeatureEngineering
feature_engineering = ChurnFeatureEngineering(data_path)
```

```python
# Display value counts for the 'churn' column before dropping outliers
feature_engineering.display_churn_value_counts()
```

```python
# Drop outliers from numeric features
feature_engineering.drop_outliers_numeric_features()
```
